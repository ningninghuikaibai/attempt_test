```python
import random
first_prize=[(4,2,1,1,1,1),(2,4,1,1,1,1),(1,1,4,2,1,1),(1,1,2,4,1,1),(1,1,1,1,4,2),(1,1,1,1,2,4)]

def roll_dice():
     return random.randint(1, 6)

def is_first_prize(data):
    for i in range(6):
        if data[i]==4 and all(x==1 for x in data[:i]+data[i+1:]):
            return True
    return False
data=[]
first_count=0
for i in range(10000):
    data.append([roll_dice() for _ in range(6)])
    if is_first_prize(data[-1]):
        first_count += 1
first_probability=first_count/10000
print("每个人摇出的数据; ")
for i in range(len(data)):
    print(data[i])
print("摇出的状元次数： ",first_count)
print("摇出的状元概率： ",first_probability)
```
