```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.randint(0,5,size=(10,10))
print('原数据'.center(50,'-'),'\n',data)
n=4
i=1+(data.shape[0]-4)
j=1+(data.shape[1]-4)
mid_data=np.lib.stride_tricks.as_strided(data,shape=(i,j,n,n),strides=data.strides+data.strides)
print('提取后的数据: '.center(50,'-'),'\n',mid_data)
```
